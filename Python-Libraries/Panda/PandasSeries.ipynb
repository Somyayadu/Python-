Pandas Series

```python
import pandas as pd
import numpy as np
```

```python
# string

country =['india','pakisthan','usa','uk','nepal']

pd.Series(country)
```

```python
# integers

runs = [13,23,45,56,34,457,84,9,0]

pd.Series(runs)
```

```python
# custom index

marks=[67,87,45,34]
subject=['math','hindi','english','sst']

pd.Series(marks,index=subject)
```

```python
pd.Series(marks,index=subject,name='somya k marks')
```

```python
# series from dictionary

marks ={
    'math':76,
    'sst':54,
    'english':95,
    'eco':91
}

marks_series=pd.Series(marks,name='somya k marks')
marks_series
```
